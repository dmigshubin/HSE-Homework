```python
from datetime import datetime, timedelta

def date_range(start_date, end_date):
    date_list = []

    #    start_date
    try:
        start = datetime.strptime(start_date, "%Y-%m-%d").date()
    except:
        print(f"{start_date}    YYYY-MM-DD")
        return date_list
        
    #    end_date
    try: 
        end = datetime.strptime(end_date, "%Y-%m-%d").date()
    except:
        print(f"{end_date}    YYYY-MM-DD")
        return date_list

    #   start_date   end_date
    if start > end:
        print(f"{start}  {end}")
        return date_list

    diff = end - start
    
    for i in range(diff.days + 1):
        date_list.append((start + timedelta(days=i)).strftime("%Y-%m-%d"))
    
    return date_list

date_range("2025-01-01", "2025-01-03")
# print(datetime.now().strftime("%Y-%m-%d"))
# print(datetime.strptime("2025-10-1", "%Y-%m-%d").date())
```
