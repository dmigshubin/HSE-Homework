```python
import json

purchases = {}

with open(file='purchase_log.txt', mode='r', encoding='utf-8') as file:
    for line in file:
        data = json.loads(line.strip())
        user_id = data['user_id']
        category = data['category']
        purchases[user_id] = category

print(list(purchases.items())[:3])
```

```python
import csv

with open(file='visit_log.csv', mode='r', encoding='utf-8') as visits, \
     open(file='funnel.csv', mode='w', newline='') as funnel:
         
         reader = csv.reader(visits)
         writer = csv.writer(funnel)
         
         #  
         writer.writerow(['user_id', 'source', 'category'])
         next(reader)
         
         #     ,   
         for user_id, source in reader:
             category = purchases.get(user_id)
             if category:
                 writer.writerow([user_id, source, category])
                 
         for i, line in enumerate(funnel):
             print(line)
             if i == 3:
                 break
```
